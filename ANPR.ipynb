# DETECTION FROM IMAGES


```python
# Step 6 + OCR: Upload ‚Üí Detect ‚Üí Extract Number
from google.colab import files
from IPython.display import Image, display
import os
import cv2
import easyocr

# Upload file
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Run YOLOv8 detection (image or video)
results = model.predict(source=filename, save=True, conf=0.5)
print(f"Detection completed on: {filename}")

# Get latest detection results folder
latest_folder = max([f for f in os.listdir('runs/detect')],
                    key=lambda x: os.path.getctime(os.path.join('runs/detect', x)))
latest_path = os.path.join('runs/detect', latest_folder)

# Display detected image
for file in os.listdir(latest_path):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        detected_image_path = os.path.join(latest_path, file)
        display(Image(filename=detected_image_path))

# ---- OCR to read license plate ----
reader = easyocr.Reader(['en'])  # English
image = cv2.imread(detected_image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Optional: Improve OCR accuracy
gray = cv2.bilateralFilter(gray, 11, 17, 17)

# Run OCR
results_ocr = reader.readtext(gray)

# Print detected text
print("\nDetected License Plate Number(s):")
for (bbox, text, prob) in results_ocr:
    print(f"{text}  (Confidence: {prob:.2f})")
```

# DETECTION FROM VIDEO

```python
# Step 7: Video Detection + OCR Overlay
!pip install easyocr opencv-python-headless moviepy

import cv2
import easyocr
from ultralytics import YOLO
from google.colab import files
from IPython.display import HTML
from base64 import b64encode
import os

# --- 1. Upload the video ---
print("Upload your test video (MP4/AVI):")
uploaded = files.upload()
video_file = list(uploaded.keys())[0]
print(f"Uploaded video: {video_file}")

# --- 2. Prepare OCR and model ---
reader = easyocr.Reader(['en'])
model = YOLO("yolov8n.pt")  # replace if you're using custom weights

# --- 3. Open video for processing ---
cap = cv2.VideoCapture(video_file)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define output video writer
output_path = "video_with_plate_text.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
writer = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_idx = 0
print("Processing video... this may take a while.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_idx += 1
    results = model(frame)  # run detection

    # For each plate detected, crop & OCR
    for r in results:
        for box in r.boxes.xyxy.cpu().numpy():
            x1, y1, x2, y2 = map(int, box)
            cropped = frame[y1:y2, x1:x2]
            gray = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
            gray = cv2.bilateralFilter(gray, 11, 17, 17)
            ocr_res = reader.readtext(gray)
            text = ocr_res[0][1] if ocr_res else ""

            # Overlay text on frame above bounding box
            if text:
                cv2.putText(frame, text, (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)

            # Draw bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

    writer.write(frame)

cap.release()
writer.release()
print(f"Finished processing. Output saved to: {output_path}")

# --- 4. Display the video in the notebook ---
mp4 = open(output_path, 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f"<video width=600 controls src='{data_url}'></video>")
```
